dataframe and series
DataFrames can be thought of as rows and columns, while a Series can be thought of as just a single column of rows.
keys are columns,values are rows.here we have 3 series first,last and email.to get rows we should use loc or iloc indexes.iloc means integer location.we can also specify particular column by iloc index after passing the index of the row.loc means searching by labels.labels for rows be as indexes.

```python
people = {
    "first" : ["corey",'jane','john'],
    "last" : ["schafer",'doe','doe'],
    "email" : ["coreymschafer@gmail.com",'janedoe@email.com','johndoe@email.com']

}
```

```python
people['email']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```
