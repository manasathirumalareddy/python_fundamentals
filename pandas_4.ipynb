filtering the dataframe.


```python
people = {
     "first" : ["corey","jane","john"],
     "last"  : ["schafer","joe","joe"],
     "emailid" : ["coreyschafer@gmail.com","janedoe@email.com","johndoe@email.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
 filt = (df['last'] == 'joe')
```

```python
df[filt]
```

```python
df.loc[filt]
```

```python
df.loc[filt,'emailid']
```

```python
filt = (df['last'] == 'joe') & (df['first'] =='john' )
```

```python
df.loc[filt,'emailid']
```

```python
filt = (df['last'] == 'schafer') | (df['first'] =='john' )
```

```python
df.loc[filt,'emailid']
```

```python
df.loc[-filt,'emailid']
```
